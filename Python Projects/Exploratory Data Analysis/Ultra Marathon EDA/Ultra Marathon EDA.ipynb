# Ultra Marathon Exploratory Data Analysis

```python
# import libraries
import numpy as np 
import pandas as pd 
import seaborn as sns 

pd.set_option('display.max_columns', 100)
```

```python
# load data
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')

df.head()
```

```python
# size
df.shape
```

```python
#types
df.dtypes
```

### Clean up data

Show 100km or 100mi event distance

```python
df[df['Event distance/length'].isin(['100km', '100mi'])]
```

```python
# add a Year Condition
df[(df['Event distance/length'].isin(['100km','100mi'])) &  (df['Year of event'].isin([2018, 2019, 2020]))]
```

```python
# find races in Ukraine
df[df['Event name'].str[-5:] == ('(UKR)')]
```

```python
# combine filters and assign result to a new dataframe
df2 = df[(df['Event name'].str[-5:] == ('(UKR)')) & (df['Event distance/length'].isin(['100km','100mi'])) &  (df['Year of event'].isin([2018, 2019, 2020]))]
df2.head()
```

```python
# remove UKR from 'Event Name'
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
print(df2.shape)
df2.head()
```

```python
# Clean up athlete age
df2['Athlete_age'] = 2024 - df2['Athlete year of birth']
df2.head()
```

```python
# Remove 'h' from athlete performance 
df2['Athlete performance'] = df2['Athlete performance'].str.split(' ').str.get(0)
df2.head()
```

```python
# Drop unnecessary columns 
df2 = df2.drop(['Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category'], axis = 1)
```

```python
df2.head()
```

**Clean null values**

```python
# find null values
df2.isna().sum()
```

```python
# look at null values
df2[df2['Athlete_age'].isna()]
```

```python
# drop null values
df2 = df2.dropna()
```

```python
# check
df2.isna().sum()
```

```python
df2.shape
```

```python
# Check for duplicates
df2[df2.duplicated() == True]
```

```python
# Reset Index
df2.reset_index(inplace = True, drop = True)
df2.head(3)
```

**Fix types**

```python
df2.dtypes
```

```python
df2['Athlete_age'] = df2['Athlete_age'].astype(int)
df2['Athlete average speed'] = df2['Athlete average speed'].astype(float)
```

```python
# Rename columns (get rid of spaces and make all in lower-case)

df2 = df2.rename(columns = {'Year of event':'year',
                        'Event dates':'race_day',
                        'Event name':'race_name',
                        'Event distance/length':'race_length',
                        'Event number of finishers':'race_number_of_finishers',
                        'Athlete performance':'athlete_performance',
                        'Athlete gender':'athlete_gender',
                        'Athlete average speed':'athlete_average_speed',
                        'Athlete ID':'athlete_id',
                        'Athlete_age':'athlete_age'
                        })
```

```python
df2.head(1)
```

```python
# Reorder columns
df3 = df2[['race_day', 
           'race_name', 
           'race_length', 
           'race_number_of_finishers', 
           'athlete_id', 'athlete_gender', 
           'athlete_age', 
           'athlete_performance', 
           'athlete_average_speed']]
df3.head(1)
```

```python
# clean race_name column
df3['race_length'] = df3['race_length'].str.rstrip('km')
df3['race_length'] = df3['race_length'].astype(int)
```

```python
# recheck data types
df3.dtypes
```

```python
df3.head()
```

```python
df3.groupby('athlete_gender')[['athlete_gender', 
                               'athlete_average_speed']].agg({'athlete_average_speed':['min', 'max', 'mean']}).sort_values(by=('athlete_average_speed', 'mean'), ascending = False)
```

```python
sns.histplot(df3, x = 'race_length', hue = 'athlete_gender')
```

`The number of male runners is significantly higher than female runners`

```python
sns.displot(df3['athlete_average_speed'],)
```

```python
df3.query('race_length == 100').groupby('athlete_age')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False).query('count>0').head(15)
```

`It's incredible, but the highest average speeds are shown by runners ages 58, 55, 43 and 53. `

### Let's take a look at how the season affects runners' speed

```python
# Extract month from date
df3['race_month'] = df3['race_day'].str.split('.').str.get(2).astype(int)

# Transform number of month into full month name 
df3['race_season'] = df3['race_month'].apply(lambda x: 'Winter' if x > 11 else 'Autumn' if x > 8 else 'Summer' if x > 5 else 'Spring' if x > 2 else 'Winter')
```

```python
# Look at the difference between speeds in different seasons
df3.groupby('race_season')['athlete_average_speed'].agg(['mean', 'count']).sort_values('mean', ascending = False)
```

**In summer and winter, runners have clearly higher speeds than in spring or fall**
